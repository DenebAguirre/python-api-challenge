```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"

Cityname = []
Country = []
Latitud = []
Longitud = []
Date = []
MaxTemperature = []
Humidity = []
WindSpeed = []
Cloudiness = []
unit = 'metric'

for city in range(len(cities)):
    print(f'Making call number {city} for {cities[city]}')
    try:
        response = requests.get(f'{url}q={cities[city]}&units={unit}&appid={weather_api_key}').json()
        Cityname.append(response['name'])
        Country.append(response['sys']['country'])
        Latitud.append(response['coord']['lat'])
        Longitud.append(response['coord']['lon'])
        Date.append(response['dt'])
        MaxTemperature.append(response['main']['temp_max'])
        Humidity.append(response['main']['humidity'])
        WindSpeed.append(response['wind']['speed'])
        Cloudiness.append(response['clouds']['all'])
    except KeyError: 
        print('Oops! Not city near here')
        pass
```

```python
#Create a dictionary with the collected data
weather_dict = {
    'City' : Cityname,
    'Country' : Country,
    'Longitud' : Longitud,
    'Latitud' : Latitud,
    'Date' : Date,
    'Max. Temperature' : MaxTemperature,
    'Humidity' : Humidity,
    'Wind Speed' : WindSpeed,
    'Cloudiness' : Cloudiness
}
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
Cities_weather = pd.DataFrame(weather_dict)
Cities_weather.to_csv('cities.csv')
Cities_weather.head()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
```

```python
# Create Northern and Southern Hemisphere DataFrames
North = Cities_weather.loc[Cities_weather['Latitud'] > 0.00]
South = Cities_weather.loc[Cities_weather['Latitud'] < 0.00]
len(South), len(North)
```

####  Northern vs South Hemisphere - Max Temp vs. Latitude Linear Regression

```python
northxvalues = North['Latitud'] 
northyvalues = North['Max. Temperature']
southxvalues = South['Latitud']
southyvalues = South['Max. Temperature']

#Here I create a general figure with commun labels for different subplots
fig5, (ax1, ax2) = plt.subplots(1, 2, figsize = (14,8))
fig5.suptitle('South vs North Max Temperature (2020-04-08 18:41:29)')
fig5.text(0.5, 0.04, 'Latitud', ha='center', va='center')
fig5.text(0.06, 0.5, 'Max. Temperature (C)', ha='center', va='center', rotation='vertical')

#Now we create the South subplot with his regression line
ax1.scatter(southxvalues, southyvalues)
(slope, intercept, rvalue, pvalue, stderr) = linregress(southxvalues, southyvalues)
regress_values = southxvalues * slope + intercept
equation = f'y ={round(slope,2)} x  + {round(intercept,2)}'
ax1.plot(southxvalues,regress_values,'r-')
ax1.annotate(equation, (-30,10), fontsize = 15, color = "red")
ax1.set_title('South Hemisphere')
ax1.set_ylim(-25,45)
ax1.set_xlim(-60,5)

#Now North
ax2.scatter(northxvalues, northyvalues)
(slope, intercept, rvalue, pvalue, stderr) = linregress(northxvalues, northyvalues)
regress_values = northxvalues * slope + intercept
equation = f'y ={round(slope,2)} x  + {round(intercept,2)}'
ax2.plot(northxvalues,regress_values,'r-')
ax2.annotate(equation, (1,1), fontsize = 15, color = "red")
ax2.set_title('North Hemisphere')
ax2.set_ylim(-25,45)
ax2.set_xlim(-5,80)

plt.savefig('South vs North Max Temperature Plot.png', bbox_inches='tight', format='png')
plt.show()
```

-With help of a regression line we can see how temperature in both hemispheres tend to increase once we get closer to the equator.
-The reason why this northern line seems to be more pronounced, is because there are more cities in northpole than in the south, but we may expect that south temperature has similar behavior
